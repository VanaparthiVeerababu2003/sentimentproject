```python
import pandas as pd

# Creating the dataframe
data = {
    "Sentiment": ["Positive", "Joy", "Excitement", "Neutral", "Contentment", "Happy", "Hopeful", "Sad", "Gratitude", "Curiosity"],
    "Count": [44, 42, 32, 14, 14, 14, 9, 9, 9, 8]
}

df1 = pd.DataFrame(data)

# Mapping categorical sentiments to numerical values (-1, 0, 1)
sentiment_mapping = {
    "Positive": 1,
    "Joy": 1,
    "Excitement": 1,
    "Contentment": 1,
    "Happy": 1,
    "Hopeful": 1,
    "Gratitude": 1,
    "Neutral": 0,
    "Curiosity": 0,
    "Sad": -1
}

# Apply mapping
df1["Sentiment_Numeric"] = df1["Sentiment"].map(sentiment_mapping)

# Now, let's separate features and target
X = df1.drop("Sentiment", axis=1)  # Dropping 'Sentiment' column
y = df1["Sentiment_Numeric"]  # Target variable with mapped sentiment values

# Displaying the target 'y'
print(y)
```

```python
# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
df=pd.read_csv("sentimentdataset.csv")
df
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

# Model building: Random Forest Classifier
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy*100)
```

```python
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```
